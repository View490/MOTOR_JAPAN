```python
model = YOLO('yolov8n-seg.pt')
```

```python
video_path = 'cars.mp4'
cap = cv2.VideoCapture(video_path)
```

```python
import cv2
import imageio
from IPython.display import Image, display
from tqdm import tqdm

# Initialize a list to store video frames
frames = []

while cap.isOpened():
    success, frame = cap.read()
    if success:
        frames.append(frame)  # Store original frames

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break

# Check if frames were captured
if len(frames) > 0:
    # Make batch predictions using your model
    annotated_frames = []
    for frame in tqdm(frames):
        results = model(frame)
        annotated_frame = results[0].plot()
        annotated_frames.append(annotated_frame)

    # Convert annotated frames to RGB format for imageio
    annotated_frames_rgb = [cv2.cvtColor(frame, cv2.COLOR_BGR2RGB) for frame in annotated_frames]

    # Save annotated frames as a GIF
    imageio.mimsave('output.gif', annotated_frames_rgb, duration=0.1)

    # Display the saved GIF
    display(Image(filename='output.gif'))
else:
    print("No frames were captured.")
```
